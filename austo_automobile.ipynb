# Problem Statement

**Q1. Do men tend to prefer SUVs more compared to women?**

```python
suv = (df[df["Make"]=="SUV"].groupby("Gender").size()/df.groupby(df["Gender"]).size()*100).round(2)
suv
```

```python
ax = sns.barplot(x = suv.index,y=suv.values,hue=suv.index,palette=['purple',"cyan"])
plt.title("Preference of SUVs")
plt.ylabel("Percentage (%)")
for i in ax.containers:
    ax.bar_label(i)
```

**Q2. What is the likelihood of a salaried person buying a Sedan?**

```python
salaried_df = df[df['Profession']=='Salaried']
salaried_sedan = salaried_df[salaried_df["Make"]=="Sedan"].shape[0]/salaried_df.shape[0]*100
salaried_other = 100-salaried_sedan
salaried_sedan = round(salaried_sedan,2)
salaried_other = round(salaried_other,2)
print(f"Likelihood of purchasing Sedan by a Salaried person is {salaried_sedan}%")
print(f"Likelihood of purchasing Sedan by a non-Salaried person is {salaried_other}%")
```

```python
ax= sns.barplot(x= ['Sedan Buyers','Other car Buyers'],y= [salaried_sedan,salaried_other],
            hue=['Sedan Buyers','Other car Buyers'],palette=['r','b'],width=0.6)
plt.title("Likelihood of Salaried Person buying a Sedan")
plt.xlabel('Car Types')
plt.ylabel("Percentage (%)")
plt.ylim(0,100)
for i in ax.containers:
    ax.bar_label(i)
```

**Q3. What evidence or data supports Sheldon Cooper's claim that a salaried male is an easier target for a SUV sale over a Sedan sale?**

```python
salaried_male = df[(df["Profession"]=="Salaried") & (df["Gender"]=="Male")]
suv = salaried_male[salaried_male["Make"]=="SUV"].shape[0] / salaried_male.shape[0]*100
sedan = salaried_male[salaried_male["Make"]=="Sedan"].shape[0] / salaried_male.shape[0]*100
suv=round(suv,2)
sedan=round(sedan,2)
print(f"Salaried Males buying a SUV: {suv}%")
print(f"Salaried Males buying a Sedan: {sedan}%")
```

```python
ax= sns.barplot(x= ['SUV','Sedan'],y= [suv,sedan],hue=['SUV','Sedan'],palette='inferno')
plt.title("Likelihood of Salaried Males buying a SUV or Sedan")
plt.xlabel('Car Types')
plt.ylabel("Percentage (%)")
plt.ylim(0,100)
for i in ax.containers:
    ax.bar_label(i)
```

**Q4. How does the the amount spent on purchasing automobiles vary by gender?**

```python
amount= df.groupby('Gender')['Price'].sum()
amount
```

```python
ax= sns.barplot(x= amount.index,y=amount.values,hue= amount.index,palette=['purple','cyan'])
plt.title("Amount Spent on Automobiles varying by Gender")
plt.xlabel('Gender')
plt.ylabel("Total Amount Spent ($)")
for i in ax.containers:
    ax.bar_label(i)
```

**Q5. How much money was spent on purchasing automobiles by individuals who took a personal loan?**

```python
loan= df[df['Personal_loan']=='Yes']
no_loan= df[df["Personal_loan"]=='No']
money_loan= loan['Price'].sum()
money_no_loan= no_loan['Price'].sum()
print(f"Money Spent by Individuals on Automobiles who took Loan: ${money_loan}")
print(f"Money Spent by Individuals on Automobiles who have No Loan: ${money_no_loan}")
```

```python
ax= sns.barplot(x= ['Yes','No'],y= [money_loan,money_no_loan],hue=['Yes','No'],palette='inferno')
plt.title("Amount spent on Automobiles by Individuals who took Personal Loan")
plt.xlabel('Personal Loan')
plt.ylabel("Amount ($)")
for i in ax.containers:
    ax.bar_label(i)
```

**Q6. How does having a working partner influence the purchase of higher-priced cars?**

```python
higher_price= df['Price'].median()
high_price_df= df[df['Price']>higher_price]
high_price_df
high_price_cars= high_price_df.groupby('Partner_working')['Price'].count()
high_price_cars
```

```python
ax= sns.barplot(x= high_price_cars.index,y=high_price_cars.values,hue= high_price_cars.index,palette='magma')
plt.title("Working Partner Influence on purchase of Higher-priced Cars")
plt.xlabel('Partner Working')
plt.ylabel("High Price Cars purchased")
plt.ylim(0,500)
for i in ax.containers:
    ax.bar_label(i)
```
